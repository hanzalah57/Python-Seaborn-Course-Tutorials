```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
titanic = sns.load_dataset("titanic")
titanic 
```

```python
sns.catplot(x='sex', y='survived', hue='class', kind='bar', data=titanic)
plt.show()
```

```python
sns.catplot(x='deck', kind='count', palette='ch:0.25', data=titanic)
plt.show()
```

```python
sns.catplot(y='deck', kind='count', hue='class', palette='pastel', edgecolor='0.6', data=titanic)
plt.show()
```

```python
sns.catplot(x='sex', y='survived', hue='class', kind='point', data=titanic)
plt.show()
```

```python
sns.catplot(x='class', y='survived', hue='sex', palette={'male':'g', 'female':'m'}, markers=['^','o'], linestyles=['-','--'], kind='point', data=titanic)
plt.show()
```

```python
iris = sns.load_dataset('iris')
iris
sns.violinplot(x=iris.species, y=iris.sepal_length, kind="box", data=iris)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,3))
sns.countplot(y='deck', color='c', data=titanic)
plt.show()
```

```python
tips = sns.load_dataset('tips')
sns.catplot(x='day', y="total_bill", hue='smoker', col='time', kind='swarm', aspect=0.6, data=tips)
plt.show()
```
