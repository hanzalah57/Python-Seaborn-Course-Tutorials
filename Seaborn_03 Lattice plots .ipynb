```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
tips = sns.load_dataset('tips')
print(tips)
```

```python
sns.set(style='darkgrid')
sns.relplot(x='total_bill', y='tip', hue='smoker', col='time', data=tips)
```

```python
fmri = sns.load_dataset('fmri')
print(fmri)
```

```python
sns.relplot(x='timepoint', y='signal', hue='subject', col='region', row='event', kind='line', height=3, estimator=None, data=fmri)
plt.show()
```

```python
sns.relplot(x='timepoint', 
            y='signal', 
            hue='event', 
            style='event',
            col='subject', 
            col_wrap=5, 
            height=3, 
            aspect=0.75,
            linewidth=2.5,
            kind='line', 
            data=fmri.query("region=='frontal'"))
plt.show()
```
