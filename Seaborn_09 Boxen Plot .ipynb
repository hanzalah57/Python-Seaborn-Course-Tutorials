```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.set(style='whitegrid')
ax = sns.boxenplot(x='tip', data=tips)
```

```python
sns.boxenplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxenplot(x='day', y='total_bill', hue='smoker', palette='Set3',data=tips)
```

```python
sns.boxenplot(x='day', y='total_bill', hue='time', linewidth=2.5, data=tips)
```

```python
sns.boxenplot(x='time', y='tip', order=['Dinner', 'Lunch'], palette='Set2', data=tips)
```

```python
#iris = sns.load_dataset('iris')
#ax = sns.boxenplot(data=iris, orient='h', palette='Set2')
#plt.show()

ax = sns.boxenplot(x='day', y='total_bill', data=tips)
ax = sns.stripplot(x='day', y='total_bill', size=4, jitter=True, color='gray', data=tips)
```

```python
g = sns.catplot(x='sex', y='total_bill', hue='smoker', col='time', aspect=0.7, height=4, kind='boxen', data=tips)
```
