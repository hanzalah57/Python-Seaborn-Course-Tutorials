```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.set(style='ticks', color_codes=True)
```

```python
sns.catplot(x='day', y='total_bill', kind='box', data=tips)
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', hue='smoker', kind='box', data=tips)
plt.show()
```

```python
tips['weekend']=tips['day'].isin(['Sat','Sun'])
```

```python
sns.catplot(x='day', y='total_bill', hue='weekend', dodge=False, kind='box', data=tips)
plt.show()
```

```python
diamonds = sns.load_dataset("diamonds")
print(diamonds)
```

```python
sns.catplot(x='color', y='price', kind='boxen', data=diamonds.sort_values("color"))
plt.show()
```

```python
sns.catplot(x='total_bill', y='day',hue='time', kind='violin', data=tips)
plt.show()
```

```python
sns.catplot(x='total_bill', y='day',hue='time', bw=0.15, cut=0, kind='violin', data=tips)
plt.show()
```

```python
sns.catplot(x='total_bill', y='day',hue='sex', split=True, kind='violin', data=tips)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill', hue='sex', split=True, inner='stick', palette='pastel', kind='violin', data=tips)
plt.show()
```

```python
g = sns.catplot(x='day', y='total_bill', inner=None, kind='violin', data=tips)
sns.swarmplot(x='day', y='total_bill', color='k', size=3, data=tips, ax=g.ax)
plt.show()
```
