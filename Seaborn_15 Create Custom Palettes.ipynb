```python
import seaborn as sns 
import matplotlib.pyplot as plt 
sns.set(style='whitegrid')
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='day',sizes=9, data=tips)
plt.grid(False)
```

# Highlighting palettes 

```python
palette_dict = {x:'gray' for x in tips.day.unique()}
palette_dict
```

```python
palette_dict['Sat'] = 'crimson'
palette_dict
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='day', palette=palette_dict, data=tips)
```

```python
sns.displot(x='tip', hue='day', col='day', palette=palette_dict, data=tips)
```
