```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.set(style='darkgrid')
sns.pointplot(x='time', y='total_bill', data=tips)
plt.show()
```

```python
sns.pointplot(x='time', y='total_bill', hue='smoker', data=tips)
plt.show()
```

```python
sns.pointplot(x='time', y='total_bill', hue='smoker', dodge=True, data=tips)
plt.show()
```

```python
ax= sns.pointplot(x='time', y='total_bill', hue='smoker', markers=['o','x'], linestyles=['-','--'], data=tips)
plt.show()
```

```python
#horizontal plot
sns.pointplot(x='tip', y='day', data=tips)
plt.show()
```

```python
#remove lines joining 
ax = sns.pointplot(x='tip', y='day', join=False, data=tips)
plt.show()
```

```python
#to use different color for each block
ax = sns.pointplot(x='tip', y='day', color='#bb3f3f',  data=tips)
plt.show()
```

```python
#to use different color paletts 
ax = sns.pointplot(x='time', y='total_bill', hue='smoker', palette='Set2',  data=tips)
plt.show()
```

```python
# to control point order
ax = sns.pointplot(x='time', y='total_bill', hue='smoker', palette='Set2', order=['Dinner', 'Lunch'],  data=tips)
plt.show()
```

```python
# compute central tendency 
import numpy as np
ax = sns.pointplot(x='time', y='total_bill', hue='smoker', palette='Set2', estimator=np.median,  data=tips)
plt.show()
```

```python
# to show stander error of mean 
ax = sns.pointplot(x='day', y='tip', ci=68, data=tips)
```

```python
# to show standard deviation of observation 
ax = sns.pointplot(x='day', y='tip', ci='sd', data=tips)
```

```python
# to enable cap size
ax = sns.pointplot(x='day', y='tip', ci='sd', capsize=0.2, data=tips)
```

```python
# combine with catplot 
g = sns.catplot(x='sex', y='total_bill', hue='smoker', col='time', kind='point', dodge=True, palette='Set2',
                height=4, aspect=0.7, data=tips)
plt.show()
```
