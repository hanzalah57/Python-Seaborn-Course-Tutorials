```python
import seaborn as sns
```

```python
sns.set(style='darkgrid')
tips = sns.load_dataset('tips')
ax = sns.barplot(x='day', y='total_bill', data=tips)
```

```python
ax = sns.barplot(x='day', hue='sex', y='total_bill', data=tips)
```

```python
ax = sns.barplot(x='tip', y='day', data=tips)
```

```python
ax = sns.barplot(x='time', y='tip', order=['Dinner','Lunch'], data=tips)
```

```python
from numpy import median 
ax = sns.barplot(x='day', y='tip', estimator=median, data=tips )
```

```python
ax = sns.barplot(x='day', y='tip', ci=68, data=tips )
```

```python
ax = sns.barplot(x='day', y='tip', ci='sd', data=tips )
```

```python
ax = sns.barplot(x='day', y='tip', capsize=0.2, data=tips )
```

```python
ax = sns.barplot(x='size', y='total_bill', palette="Blues_d", data=tips )
```

```python
tips["weekend"] = tips["day"].isin(["Sat","Sun"])
sns.barplot(x='day', y='total_bill', hue='weekend', data=tips, dodge=False)
```

```python
ax = sns.barplot(x='size', y='total_bill', data=tips, color='salmon', saturation=0.5)
```

```python
ax = sns.barplot(x='day', y='total_bill', data=tips, linewidth=2.5, facecolor=(1,1,1,0), errcolor='0.2', edgecolor='0.2')
```

```python
g = sns.catplot(x='sex', y='total_bill', hue='smoker', col='time', data=tips, kind='bar', height=4, aspect=0.7 )
```
