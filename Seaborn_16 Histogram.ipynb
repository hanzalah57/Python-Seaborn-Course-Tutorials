```python
import seaborn as sns 
```

```python
sns.__version__
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
sns.set(style='darkgrid')
sns.histplot(penguins.bill_depth_mm)
```

```python
sns.histplot(y='bill_depth_mm', data=penguins)
```

```python
sns.histplot(x='bill_depth_mm', kde=True, data=penguins)
```

```python
sns.histplot(x='bill_depth_mm', kde=True, bins=20, data=penguins)
```

```python
sns.histplot(x='bill_depth_mm', kde=True, binwidth=2, data=penguins)
```

```python
sns.histplot(x='bill_depth_mm', kde=True, binwidth=2, binrange=[16,20], data=penguins)
```

```python
sns.histplot(x='bill_depth_mm', stat='count', data=penguins)
```

```python
sns.histplot(x='bill_depth_mm', stat='density', data=penguins)
```

```python
sns.histplot(x='bill_depth_mm', stat='probability', data=penguins)
```

```python
sns.histplot(x='bill_depth_mm', stat='probability', fill=False, data=penguins)
```

```python
sns.histplot(x='bill_depth_mm', stat='probability', fill=False, element='step', data=penguins)
```

```python
sns.histplot(x='bill_depth_mm', stat='probability', fill=False, element='step', cumulative=True, data=penguins)
```

```python
sns.histplot(x='bill_depth_mm', hue='species', data=penguins)
```

```python
sns.histplot(x='bill_depth_mm', hue='species', element='step', data=penguins)
```

```python
sns.histplot(x='bill_depth_mm', hue='species', element='poly', data=penguins)
```

```python
sns.histplot(x='bill_depth_mm', hue='species', multiple='layer', data=penguins)
```

```python
sns.histplot(x='bill_depth_mm', hue='species', multiple='stack', data=penguins)
```

```python
sns.histplot(x='bill_depth_mm', hue='species', multiple='fill', data=penguins)
```
