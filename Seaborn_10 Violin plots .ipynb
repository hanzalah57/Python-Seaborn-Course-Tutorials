```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
tips = sns.load_dataset("tips")
sns.set(style="whitegrid")
```

```python
ax = sns.violinplot(x='total_bill', data=tips)
```

```python
ax = sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
ax = sns.violinplot(x='day', y='total_bill', hue='smoker', palette='muted', data=tips)
```

```python
ax = sns.violinplot(x='day', y='total_bill', hue='smoker', palette='muted', split=True, data=tips)
```

```python
ax = sns.violinplot(x='time', y='tip', order=['Dinner', 'Lunch'], data=tips)
```

```python
ax = sns.violinplot(x='day', y='total_bill', hue='sex', palette='Set2', split=True, scale='count', data=tips)
```

```python
ax = sns.violinplot(x='day', y='total_bill', hue='sex', palette='Set2', split=True, scale='count', inner='quartile', data=tips)
```

```python
ax = sns.violinplot(x='day', y='total_bill', hue='sex', palette='Set2', split=True, scale='count', inner='stick', data=tips)
```

```python
ax = sns.violinplot(x='day', y='total_bill', hue='sex', palette='Set2', spli=True, scale='count', inner='stick', scale_hue=False,
                    data=tips)
```

```python
ax = sns.violinplot(x='day', y='total_bill', hue='sex', palette='Set2', spli=True, scale='count', inner='stick', scale_hue=False,
                    bw=0.2, data=tips)
```

```python
planets = sns.load_dataset("planets")
planets
```

```python
ax = sns.violinplot(x='orbital_period', y='method', data=planets[planets.orbital_period < 1000], scale='width', palette="Set3")
```

```python
ax = sns.violinplot(x='orbital_period', y='method', data=planets[planets.orbital_period < 1000], cut=0, scale='width', palette="Set3")
```

```python
tips["weekend"] = tips["day"].isin(["Sat","Sun"])
sns.violinplot(x='day', y='total_bill', hue='weekend', data=tips, dodge=True)
```

```python
g = sns.catplot(x='sex', y='total_bill', hue='smoker', col='time', split=True, kind='violin', height=4, aspect=0.7, data=tips)
```
