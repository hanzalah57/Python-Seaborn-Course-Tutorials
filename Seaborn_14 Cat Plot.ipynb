```python
import seaborn as sns 
sns.set(style='ticks')
exercise = sns.load_dataset('exercise')
```

```python
ax = sns.catplot(x='time', y='pulse', hue='kind', data=exercise)
```

```python
ax = sns.catplot(x='time', y='pulse', hue='kind', data=exercise, kind='violin')
```

```python
ax = sns.catplot(x='time', y='pulse', hue='kind', col='diet', data=exercise)
```

```python
ax = sns.catplot(x='time', y='pulse', hue='kind', col='diet', data=exercise, height=5, aspect=0.8)
```

```python
titanic = sns.load_dataset('titanic')
sns.catplot(x='alive', col='deck', col_wrap=4, data=titanic[titanic.deck.notnull()], height=2.5, aspect=0.8, kind='count')
```

```python
g = sns.catplot(x='age', y='embark_town', hue='sex', row='class', data=titanic[titanic.embark_town.notnull()],
               orient='h', palette='Set3', height=2, aspect=3, kind='violin', dodge=True, cut=0, bw=0.2)
```

```python
g = sns.catplot(x='who', y='survived', col='class', data=titanic, saturation=0.5, kind='bar', errorbar=None, aspect=0.6)
g.set_axis_labels("","Survival Rate")
g.set_xticklabels(['Men', 'Women', 'Children'])
g.set_titles("{col_name},{col_var}")
g.set(ylim=(0,1))
```
