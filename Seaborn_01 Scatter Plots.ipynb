```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
import lasio
```

```python
las=lasio.read('C:/Users/Samsung/Desktop/Meyal 17.las')
df=las.df()
df.describe()
```

```python
df.head()
```

```python
sns.set(style='darkgrid')
sns.relplot(x=df['TNPH'].values, 
            y=df['RHOB'].values, 
            hue=df['GR'].values, 
            size=df['GR'].values,
            legend='brief', 
            palette="ch:r=-5,l=0.75")

# Change the X and Y ranges
plt.xlim(0, 0.8)

# For the y axis, we need to flip by passing in the scale values in reverse order
plt.ylim(3.0, 1.2)


plt.ylabel('Bulk Density (DEN) - g/cc', fontsize=14)
plt.xlabel('Neutron Porosity (NEU) - %', fontsize=14)
plt.title("Porosity-Density Cross plot", fontsize=14, fontweight='bold')



plt.show()
```

```python
sns.set(style='darkgrid')
sns.relplot(x=df['DT'].values, y=df['RHOB'].values, hue=df['GR'].values, size=df['GR'].values, palette="ch:r=-5,l=0.75")
# Change the X and Y ranges
#plt.xlim(0, 0.8)

# For the y axis, we need to flip by passing in the scale values in reverse order
plt.ylim(3.0, 1.2)


plt.ylabel('Bulk Density (DEN) - g/cc', fontsize=14)
plt.xlabel('DT us/ft', fontsize=14)



plt.show()
```
