```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
sns.set(style='whitegrid')
tips = sns.load_dataset('tips')
```

```python
sns.stripplot(x='total_bill', data=tips)
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', palette='Set2', data=tips)
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', palette='Set2', data=tips, jitter=True)
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', palette='Set2', data=tips, jitter=0.05)
plt.show()
```

```python
sns.stripplot(y='day', x='total_bill', data=tips, jitter=True)
plt.show()
```

```python
sns.stripplot(y='day', x='total_bill', data=tips, jitter=True, linewidth=1)
plt.show()
```

```python
sns.stripplot(x='sex', y='total_bill', hue='day', data=tips, jitter=True)
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', hue='smoker', data=tips, jitter=True, palette='Set2', dodge=True)
plt.show()
```

```python
sns.stripplot(x='time', y='tip', data=tips, order=['Dinner', 'Lunch'])
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', hue='smoker', palette='Set2', size=20, marker='D', edgecolor='gray', alpha=0.25, data=tips)
plt.show()
```

```python
import numpy as np
ax = sns.boxplot(x='tip', y='day', data=tips, whis=np.inf)
ax = sns.stripplot(x='tip', y='day', data=tips, jitter=True, color='0.3')
```

```python
ax = sns.violinplot(x='day', y='total_bill', data=tips, inner=None, color='0.8')
ax = sns.stripplot(x='day', y='total_bill', data=tips, palette="Set2", jitter=True)
```

```python
g = sns.catplot(x='sex', y='total_bill', hue='smoker', col='time', data=tips, kind='strip', jitter=True, height=4, aspect=0.7)
```
