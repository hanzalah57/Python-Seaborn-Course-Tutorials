```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
sns.set(style='whitegrid')
tips = sns.load_dataset('tips')
```

```python
ax = sns.swarmplot(x='total_bill', data=tips)
```

```python
ax = sns.swarmplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
ax = sns.swarmplot(y='day', x='total_bill', data=tips)
plt.show()
```

```python
ax = sns.swarmplot(x='day', y='total_bill', hue='sex', data=tips)
plt.show()
```

```python
ax = sns.swarmplot(x='day', y='total_bill', hue='smoker', palette='Set2', dodge=True, data=tips)
plt.show()
```

```python
ax = sns.swarmplot(x='time', y='tip', order=['Dinner', 'Lunch'], data=tips)
plt.show()
```

```python
ax = sns.swarmplot(x='time', y='tip', order=['Dinner', 'Lunch'], size=6, data=tips)
plt.show()
```

```python
import numpy as np
ax = sns.boxplot(x='tip', y='day', data=tips, whis=np.inf)
ax = sns.swarmplot(x='tip', y='day', data=tips, color='0.2')
plt.show()
```

```python
ax = sns.violinplot(x='day', y='total_bill', data=tips, inner=None)
ax = sns.swarmplot(x='day', y='total_bill', data=tips, color='white', edgecolor='gray')
```

```python
g = sns.catplot(x='sex', y='total_bill', data=tips, hue='smoker', col='time', kind='swarm', height=4, aspect=0.7)
```
