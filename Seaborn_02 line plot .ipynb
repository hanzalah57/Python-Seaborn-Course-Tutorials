```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 
```

```python
df = pd.DataFrame(dict(time=np.arange(500), value=np.random.randn(500).cumsum()))
df.head()
df.describe()
```

```python
sns.set(style='darkgrid')

g = sns.relplot(x='time', y='value', kind='line', data=df)
g.fig.autofmt_xdate()
plt.show()
```

```python
df = pd.DataFrame(np.random.randn(500,2).cumsum(axis=0), columns=['x','y'])
sns.relplot(x='x', y='y', sort=False, kind='line', data=df)
plt.show()
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(x='timepoint', y='signal', kind='line', data= fmri, color='#ff0066')
plt.show()
```

```python
sns.relplot(x='timepoint', 
            y='signal', 
            ci=None, 
            kind='line', 
            data= fmri, 
            lw=3,
            color='#ff0066')
plt.show()
```

```python
sns.relplot(x='timepoint', 
            y='signal', 
            ci='sd', 
            kind='line', 
            data= fmri, 
            lw=3,
            color='#ff0066')
plt.show()
```

```python
sns.relplot(x='timepoint', 
            y='signal', 
            estimator=None, 
            kind='line', 
            data= fmri, 
            lw=3,
            color='#ff0066')
plt.show()
```

```python
sns.relplot(x='timepoint', 
            y='signal', 
            hue='event', 
            kind='line', 
            data= fmri, 
            lw=3,)
plt.show()
```

```python
sns.relplot(x='timepoint', 
            y='signal', 
            hue='region', 
            kind='line',
            style='event',
            data= fmri, 
            lw=3,
            color='#ff0066')
plt.show()
```

```python
sns.relplot(x='timepoint', 
            y='signal', 
            hue='region',
            dashes=False,
            markers=True,
            kind='line',
            style='event',
            data= fmri, 
            lw=3,
            color='#ff0066')
plt.show()
```

```python
sns.relplot(x='timepoint', 
            y='signal', 
            hue='event',
            style='event',
            kind='line',
            data= fmri)
plt.show()
```

```python
sns.relplot(x='timepoint', 
            y='signal', 
            hue='region',
            units='subject', 
            estimator=None,
            kind='line',
            data= fmri.query("event=='stim'"))
plt.show()
```

```python
dots = sns.load_dataset('dots').query("align=='dots'")
dots
```

```python
sns.relplot(x='time',
            y='firing_rate',
            hue='coherence',
            style='choice',
            kind='line',
            data=dots)
plt.show()
```

```python
palette = sns.cubehelix_palette(light=0.8, n_colors=6)
sns.relplot(x='time',
            y='firing_rate',
            hue='coherence',
            style='choice',
            kind='line',
            palette=palette,
            data=dots)
plt.show()
```

```python
from matplotlib.colors import LogNorm
palette = sns.cubehelix_palette(light=0.7, n_colors=6)
sns.relplot(x='time',
            y='firing_rate',
            hue='coherence',
            style='choice',
            kind='line',
            hue_norm=LogNorm(),
            data=dots)
plt.show()
```

```python
sns.relplot(x='time',
            y='firing_rate',
            size='coherence',
            style='choice',
            kind='line',
            data=dots)
plt.show()
```
